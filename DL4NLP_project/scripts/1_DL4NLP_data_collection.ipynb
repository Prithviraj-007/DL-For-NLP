```python
# Function to download image from URL and convert it to numpy array
def download_and_convert_to_array(url):
    response = requests.get(url)
    if response.status_code == 200:
        # Convert image content to numpy array
        nparr = np.frombuffer(response.content, np.uint8)
        # Read the array as an image using OpenCV
        img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
        return img
    else:
        return None
```

```python
# Apply the function to download images and store numpy arrays in a new column
final_data_df['image_array'] = final_data_df['primaryImage'].apply(download_and_convert_to_array)
```

```python
# Function to extract the description from a given URL
def extract_description(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            # Parse the HTML content of the page
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find the element containing the description
            description_element = soup.find('div', class_='artwork__intro__desc')

            # Extract the description text
            if description_element:
                description = description_element.get_text(strip=True)
                return description
            else:
                return None
        else:
            return None
    except Exception as e:
        return str(e)
```

```python
# create new column with description
final_data_df['description'] = final_data_df['objectURL'].apply(extract_description)
```
